## Define the interval bounds which contain 99% of the observations   

> **Hint**: Recall the empirical rule related to $3\sigma$.

```python
# Perform any calculations necessary here
mean = data.iloc[:, 0].mean()
std_dev = data.iloc[:, 0].std()

mean, std_dev
```

```python
# Write your answer here 
lower_bound = mean - (3 * std_dev)
upper_bound = mean + (3 * std_dev)

lower_bound, upper_bound
```

## Compute and interpret the z-score for a tree yielding 35 pounds of apples

```python
# Calculate z

z = (35 - mean) / std_dev
z
```

```python
# Interpret the result
```

## Interpretation:
**Negative z-score**: The z-score is negative, which indicates that the yield of 35 pounds is below the mean yield of the dataset.

**Magnitude:** The z-score of -1.23 means that 35 pounds is 1.23 standard deviations below the mean yield.

This suggests that a tree yielding 35 pounds of apples is somewhat below average, but not extremely uncommon in this dataset. In a normal distribution, a z-score of -1.23 falls within the range where you would expect to find a reasonable proportion of the data (between -2 and 2 standard deviations from the mean).

##  Convert each tree’s yield to a z-score so the new variable is the “z-score for weight”

The units are still the apple trees. For the data set of all z-scores:

* What is the shape? 
* The mean? 
* The standard deviation?

```python
# Give your solution here 
data['z_score'] = (data.iloc[:, 0] - mean) / std_dev

plt.figure(figsize=(10, 6))
plt.hist(data['z_score'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Z Score')
plt.xlabel('Z Score')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

1. **Shape**: The shape of the distribution of z-scores will be the same as the shape of the original data distribution. If the original data is approximately normally distributed, the z-scores will also be normally distributed. However, the z-scores themselves will always have the properties of a standard normal distribution (mean of 0 and standard deviation of 1) regardless of the original data's distribution shape.

2. **Mean**: The mean of the z-scores is always 0. This is because the z-score transformation standardizes the data around the mean.

3. **Standard Deviation**: The standard deviation of the z-scores is always 1. This standardization is done by dividing by the original data's standard deviation
